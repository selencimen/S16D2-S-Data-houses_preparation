```python
none_cols = [
    "Alley", "PoolQC", "Fence", "MiscFeature",
    "FireplaceQu",
    "GarageType", "GarageFinish", "GarageQual", "GarageCond",
    "BsmtQual", "BsmtCond", "BsmtExposure",
    "BsmtFinType1", "BsmtFinType2"
]

for col in none_cols:
    df[col] = df[col].fillna("None")
```

```python
median_cols = [
    "LotFrontage",
    "MasVnrArea",
    "GarageYrBlt",
    "BsmtFinSF1", "BsmtFinSF2",
    "BsmtUnfSF", "TotalBsmtSF"
]

for col in median_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
mode_cols = [
    "MasVnrType",
    "Electrical"
]

for col in mode_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df.isnull().sum().sort_values(ascending=False).head(10)
```

```python
df = df.drop(columns=["WallMat"])
```

```python
df["RoofSurface"] = df["RoofSurface"].fillna(df["RoofSurface"].mode()[0])
```

```python
df = df.dropna(subset=["Pesos"])
```

```python
df.isnull().sum().sort_values(ascending=False).head(10)
```

```python
ordinal_map = {
    "Po": 1,
    "Fa": 2,
    "TA": 3,
    "Gd": 4,
    "Ex": 5,
    "None": 0
}

ordinal_cols = [
    "ExterQual", "ExterCond",
    "BsmtQual", "BsmtCond",
    "HeatingQC", "KitchenQual",
    "FireplaceQu", "GarageQual", "GarageCond", "PoolQC"
]

for col in ordinal_cols:
    if col in df.columns:
        df[col] = df[col].map(ordinal_map)
```

```python
binary_cols = ["CentralAir", "PavedDrive"]

binary_map = {
    "Y": 1,
    "N": 0,
    "P": 0
}

for col in binary_cols:
    df[col] = df[col].map(binary_map)
```

```python
nominal_cols = df.select_dtypes(include="object").columns
nominal_cols
```

```python
df = pd.get_dummies(df, columns=nominal_cols, drop_first=True)
```

```python
df["HouseAge"] = df["YrSold"] - df["YearBuilt"]
```

```python
df["TotalSF"] = (
    df["TotalBsmtSF"] +
    df["1stFlrSF"] +
    df["2ndFlrSF"]
)
```

```python
df["BathScore"] = (
    df["FullBath"] +
    0.5 * df["HalfBath"] +
    df["BsmtFullBath"] +
    0.5 * df["BsmtHalfBath"]
)
```

```python
import numpy as np

df["LogPesos"] = np.log1p(df["Pesos"])
```

```python
df = df.drop(columns=["Pesos"])
```

```python
df.shape
df.head()
```

```python
X = df.drop(columns=["LogPesos"])
y = df["LogPesos"]
```

```python
from sklearn.preprocessing import StandardScaler

num_cols = X.select_dtypes(include=["int64", "float64"]).columns

scaler = StandardScaler()
X[num_cols] = scaler.fit_transform(X[num_cols])
```

```python
X.shape
```

```python
from sklearn.linear_model import Lasso

lasso = Lasso(alpha=0.001, random_state=42, max_iter=10000)
lasso.fit(X, y)

lasso_coef = pd.Series(lasso.coef_, index=X.columns)
selected_features = lasso_coef[lasso_coef != 0].index

len(selected_features)
```

```python
X_selected = X[selected_features]
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.inspection import permutation_importance

rf = RandomForestRegressor(
    n_estimators=200,
    random_state=42,
    n_jobs=-1
)

rf.fit(X_selected, y)

perm = permutation_importance(
    rf,
    X_selected,
    y,
    n_repeats=10,
    random_state=42,
    n_jobs=-1
)

importances = pd.Series(
    perm.importances_mean,
    index=X_selected.columns
).sort_values(ascending=False)

importances.head(10)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()

rmse_base = -cross_val_score(
    lin_reg,
    X_selected,
    y,
    cv=5,
    scoring="neg_root_mean_squared_error"
).mean()

rmse_base
```

```python
from sklearn.ensemble import GradientBoostingRegressor

gbr = GradientBoostingRegressor(
    n_estimators=300,
    learning_rate=0.05,
    max_depth=3,
    random_state=42
)

rmse_gbr = -cross_val_score(
    gbr,
    X_selected,
    y,
    cv=5,
    scoring="neg_root_mean_squared_error"
).mean()

rmse_gbr
```

## Final Summary

Bu √ßalƒ±≈ümada, 84 feature i√ßeren Ames Housing tabanlƒ± ev fiyatlarƒ± veri seti √ºzerinde
u√ßtan uca bir feature engineering ve feature selection s√ºreci uygulanmƒ±≈ütƒ±r.

### 1. Veri Hazƒ±rlama & Preprocessing
- Eksik deƒüerler, deƒüi≈ükenlerin anlamƒ±na g√∂re ele alƒ±nmƒ±≈ütƒ±r:
  - Yapƒ±sal olarak ‚Äúyok‚Äù anlamƒ±na gelen deƒüi≈ükenler `"None"` ile doldurulmu≈ütur.
  - Sayƒ±sal deƒüi≈ükenler median, kategorik deƒüi≈ükenler mode ile doldurulmu≈ütur.
  - Bilgi ta≈üƒ±mayan ve %99‚Äôdan fazla eksik deƒüere sahip kolonlar (√∂rn. `WallMat`) veri setinden √ßƒ±karƒ±lmƒ±≈ütƒ±r.
  - Hedef deƒüi≈ükende (`Pesos`) eksik deƒüeri olan satƒ±rlar veri setinden kaldƒ±rƒ±lmƒ±≈ütƒ±r.

### 2. Encoding
- Ordinal deƒüi≈ükenler (√∂rn. kalite skorlarƒ±) anlamlƒ± sƒ±ralamaya g√∂re sayƒ±sal olarak encode edilmi≈ütir.
- Binary deƒüi≈ükenler 0/1 formatƒ±na d√∂n√º≈üt√ºr√ºlm√º≈üt√ºr.
- Nominal deƒüi≈ükenler One-Hot Encoding y√∂ntemiyle modellenmi≈ütir.

### 3. Feature Engineering
Model performansƒ±nƒ± artƒ±rmak amacƒ±yla yeni deƒüi≈ükenler √ºretilmi≈ütir:
- `HouseAge`: Evin satƒ±≈ü yƒ±lƒ±ndaki ya≈üƒ±
- `TotalSF`: Toplam ya≈üam alanƒ± (bodrum + katlar)
- `BathScore`: Banyo sayƒ±larƒ±nƒ±n aƒüƒ±rlƒ±klƒ± birle≈üimi

Hedef deƒüi≈üken saƒüa √ßarpƒ±k daƒüƒ±lƒ±m g√∂sterdiƒüi i√ßin `LogPesos` d√∂n√º≈ü√ºm√º uygulanmƒ±≈ütƒ±r.

### 4. Scaling
- Sayƒ±sal deƒüi≈ükenler `StandardScaler` ile √∂l√ßeklendirilmi≈ütir.
- One-hot encoded deƒüi≈ükenler √∂l√ßekleme dƒ±≈üƒ±nda tutulmu≈ütur.

### 5. Feature Selection
- Y√ºksek boyutlu feature uzayƒ±nƒ± daraltmak i√ßin **Lasso Regression** kullanƒ±lmƒ±≈ütƒ±r.
- Se√ßilen feature‚Äôlar **Permutation Importance** ile doƒürulanmƒ±≈ütƒ±r.

### 6. Modelleme
- Base model olarak Linear Regression kurulmu≈ütur.
- Daha g√º√ßl√º bir model olarak Gradient Boosting Regressor kullanƒ±lmƒ±≈ütƒ±r.
